```python
df_blankage = df[df['Age'].isna()]
```

```python
#convert embarked into cataegory
df["Embarked"] = df["Embarked"].astype("category")

#convert sex into category

df['Sex'] = df['Sex'].astype("category")
```

```python
df.info()
```

```python
df['Name']
# Look at the pattern, its sir name first, then title. first name.
```

```python
#lets get their title out from their name
# PATTERN : Surname[,] TITLE[.] First Name (not important)
df['Title'] = df['Name'].str.split('[,.]').str.get(1)
```

```python
df['Surname'] = df['Name'].str.split('[,.]').str.get(0)
```

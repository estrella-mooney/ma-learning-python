## An XY Scatter Plot

An XY scatter plot puts one dot on the graph for each data point.

The `graphing` package provides this functionality in just one line.

If the data have more than two columns, it will make a guess at which to graph.

We want to graph the `number_of_workers` on the x-axis and the `tons_of_beans_harvested` on the y-axis.

Let's try to make a graph now

### Instructions (1 of 2)
Write `graphing.scatter_2D(data)` in the cell below and run it to create your graph.

```python
# Write your code here:
graphing.scatter_2D(data)
```

As you can see, `graphing` will automatically show all graphs, unless told otherwise.

Also, it hasn't picked the columns we want. Let's tell it which columns we want.

### Instructions (2 of 2)
1. Replace the first question marks with the name of the column we want on the x-axis
1. Replace the second question marks with the name of the column we want on the y-axis

### Hints
* Remember your column names need to be in quotes "".
* We want the number of workers on the x-axis and the amount of beans harvested on the y axis
* You can see the exact column names by looking at the table you printed earlier

```python
graphing.scatter_2D(data, label_x="number_of_workers", label_y="tons_of_beans_harvested")
```

## Bar Chart

We don't use bar charts very often in ML, but let's make one now just to take a look at things being made without `graphing`.

If we don't need much customisation, bar charts are quite easy to make.

### Instructions
1. Take a quick look at the code below. See if you can make sense of it.
1. Run the cell
1. Hover your mouse over the cell. Notice how it will tell you values you are hovering over.

```python
# Import the package called plotly express
import plotly.express

# Make an object that is a graph
# Our data is the dataframe from before
# We want Year on the X axis
# We want number_of_workers on the Y axis
fig = plotly.express.bar(data, x='Year', y='number_of_workers')

# Show the graph
fig.show()
```
